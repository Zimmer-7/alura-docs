```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
import plotly.express as px
```

```python
db = pd.read_csv('/content/imigrantes_canada.csv')
db
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
db.set_index('País', inplace=True)
db
```

```python
brasil = db.loc['Brasil', anos]
peru = db.loc['Peru', anos]
argentina = db.loc['Argentina', anos]
```

```python
brasil_dict = {
    'ano': brasil.index.tolist(),
    'imigrantes': brasil.values.tolist()
}

dadosBrasil = pd.DataFrame(brasil_dict)

peru_dict = {
    'ano': peru.index.tolist(),
    'imigrantes': peru.values.tolist()
}

dadosPeru = pd.DataFrame(peru_dict)

argentina_dict = {
    'ano': argentina.index.tolist(),
    'imigrantes': argentina.values.tolist()
}

dadosArgentina = pd.DataFrame(argentina_dict)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15,5))
ax[0].plot(dadosBrasil['ano'], dadosBrasil['imigrantes'])
ax[0].xaxis.set_major_locator(plt.MultipleLocator(5))
ax[0].set_xlabel('anos')
ax[0].set_ylabel('imigrantes')
ax[0].set_title('imigrantes Brasil/Canadá')
ax[0].grid()

ax[1].boxplot(dadosBrasil['imigrantes'])
ax[1].set_ylabel('imigrantes')
ax[1].set_title('imigrantes Brasil/Canadá')
ax[1].grid()

plt.show()
```

```python
top10 = db.sort_values('Total', ascending=False).head(10)
```

```python
fig = px.line(dadosBrasil, x='ano', y='imigrantes',
              title='Imigração Brasil/Canadá 1980 a 2013')

fig.update_layout(font_family = 'Times New Roman',
                  font_size = 12,
                  font_color = 'gray',
                  xaxis_title = 'Ano',
                  yaxis_title = 'Numero de imigrantes')

fig.show()
```

```python
fig.write_html('imigracao brasil_Canada')
```

```python
fig, axs = plt.subplots(figsize=(8,4))
fig.suptitle('Imigrações para o Canadá')

axs.plot(db.loc['Brasil', anos])

axs.plot(db.loc['Argentina', anos])

axs.plot(db.loc['Colômbia', anos])

axs.plot(db.loc['Peru', anos])

axs.xaxis.set_major_locator(plt.MultipleLocator(5))
axs.set_xlabel('anos')
axs.set_ylabel('imigrantes')
axs.grid()

axs.set_ylim(0,7000)

plt.show()
```
